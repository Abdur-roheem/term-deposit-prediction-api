```python
categorical = ['job', 'marital', 'education', 'default',  'housing',
               'loan', 'contact', 'month', 'poutcome']

numerical = ['age', 'balance', 'day', 'duration', 
             'campaign', 'pdays','previous']
```

```python
print(df[numerical].dtypes)
print()
print(df.y.unique())
```

```python
df[numerical].corr()
```

```python
# Convert 'yes'/'no' to 1/0
df['y'] = df['y'].map({'yes': 1, 'no': 0})

# Now try the correlation again
df[numerical].corrwith(df.y).sort_values(ascending=False)
```

duration has like 40% correlation with the target variable


```python
num_columns = df[numerical]
num_columns = num_columns.drop(columns=['balance','duration'])

num_columns.boxplot()
plt.show()
```

```python
df[numerical].hist(
    bins=20,
    figsize=(12, 4 * 3),
    layout=(4, 2)
)

plt.tight_layout()
plt.show()
```

```python
# Correlation Plot

# Compute correlation matrix
corr = df[numerical].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", cbar=True, square=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
from sklearn.metrics import mutual_info_score
```

```python
def mutual_info_churn_score(series):
    return mutual_info_score(series, df.y)
```

```python
mi = df[categorical].apply(mutual_info_churn_score)
mi.sort_values(ascending=False)
```
